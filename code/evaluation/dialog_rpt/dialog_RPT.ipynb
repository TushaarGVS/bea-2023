```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
dialogRPT_upvotes = np.load(
    f'data{os.sep}train_with-reference{os.sep}dialogue-RPT-microsoft-DialogRPT-updown.npy')
```

```python
plt.figure(dpi=200)
plt.hist(dialogRPT_upvotes, bins=20)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Avg Likelihood of Upvotes', fontsize=12)
plt.title('DialogRPT on Training Set', fontsize=14)
```

```python
dialogRPT_upvotes.mean()
```

```python
dialogRPT_upvotes.std(ddof=1)
```

```python
np.median(dialogRPT_upvotes)
```

```python
np.quantile(dialogRPT_upvotes, 0.75)
```
